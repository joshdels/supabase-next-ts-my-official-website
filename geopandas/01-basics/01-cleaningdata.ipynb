```python
import geopandas as gpd
```

```python
gdf = gpd.read_file("/home/joshua/Downloads/world.json")
print(gdf.columns)
```

```python
gdf_filter = gdf[["admin", "adm0_a3", "geometry"]]
gdf_clean = gdf_filter.rename(columns={
    "admin": "country_value",
    "adm0_a3": "county_id"
})
print(gdf_clean)
```

```python
gdf_clean.to_file("../data/world_clean.geojson", driver="GeoJSON")
```
